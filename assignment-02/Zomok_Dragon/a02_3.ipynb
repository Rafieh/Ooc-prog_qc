```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
import math
```

```python
def animate_ry(phi, steps=3):
    circuit = QuantumCircuit(steps+1, steps+1)
    #ircuit.rx(0, 0)
    temp_phi = 0
    for counter in range(steps+1):
        circuit.ry(temp_phi, counter)
        temp_phi += phi/steps
    simulator = Aer.get_backend('statevector_simulator')
    result = execute(circuit, backend=simulator).result()
    statevector = result.get_statevector()
    return plot_bloch_multivector(statevector)
```

```python
animate_ry(0.5*pi)
```

```python
phi = 0.4*pi
steps = 4
circuit = QuantumCircuit(steps, steps)
states = []
for _ in range(steps):
    circuit.rx(phi/steps, _)
    #circuit.rx(phi, 0)
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
statevector = result.get_statevector()
%matplotlib inline
plot_bloch_multivector(statevector)
```
